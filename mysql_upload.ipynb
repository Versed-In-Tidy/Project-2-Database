```python
import pandas as pd
names = pd.read_csv('names.csv')
academics = pd.read_csv('academics.csv')
music = pd.read_csv('music.csv')
clubs = pd.read_csv('clubs.csv')
```

```python
names.to_sql(name='names', con=cnx, if_exists='append', index=False)
academics.to_sql(name='academics', con=cnx, if_exists='append', index=False)
music.to_sql(name='music', con=cnx, if_exists='append', index=True)
clubs.to_sql(name='clubs', con=cnx, if_exists='append', index=True)
```

```python
pd.read_sql('''SELECT * FROM names 
               JOIN music ON names.fav_artist = music.artist_name
               JOIN academics ON names.fav_class = academics.course_id
               JOIN clubs ON names.fav_club = clubs.index
                   WHERE state = "VA" AND hometown = "Virginia Beach"''', con = cnx)
```

```python
pd.read_sql('''SELECT genre, platinum_albums,
               SUM(platinum_albums)
               FROM music
               GROUP BY genre;''', con = cnx)
```
